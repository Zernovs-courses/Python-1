# Генераторы

```
generator
```

1. содержит выражение `yield`
2. при вызове возвращает объект типа генератор, но не начинает выполнение
3. после вызова функция приостанавливается, а управление передается вызывающей стороне
4. локальные переменные и их состояние запоминаются между последовательными вызовами
5. при завершении вызывается `StopIteration`

````python
[1, 2, 3, 4].__sizeof__()
````

````python
def counter():
    val = 0
    while True:
        yield val
        val += 1

c = counter()
type(c)
````

````python
next(c)
````

````python
for i in c:
    print(i)
    if i > 100:
        break
````

````python
def primes(n):
    for i in range(2, n):
        for j in range(2, int(i ** 0.5) + 1):
            if i % j == 0:
                break
        else:
            yield i


p = primes(10000)

for prime in p:
    print(prime)
````

## Ограничения генераторов

1. генераторы можно только создать заново, но не вернуться к предыдущему состоянию
2. генераторы не поддерживают получение длины (`len`)
3. не поддерживают получение элемента по индексу
4. к ним нельзя применять операции среза
5. после использования/итерации по генератору, он остается пустым

## `yield from <expression>`

````python
def gen_list1(iterable):
    for i in list(iterable):
        yield i

def gen_list2(iterable):
    yield from list(iterable)

print(list(gen_list1("Sirius")))
print(list(gen_list2("Sirius")))
````

````python
def accumulate():
    tally = 0
    while 1:
        next = yield
        if next is None:
            return tally
        tally += next

def gather(tallies):
    while 1:
        tally = yield from accumulate()
        tallies.append(tally)

tallies = []
acc = gather(tallies)

next(acc)
for i in range(4):
    acc.send(i)

acc.send(None)
for i in range(5):
    acc.send(i)

acc.send(None)
tallies
````
