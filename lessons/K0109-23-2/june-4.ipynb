```python
memo = set()

def f(arr, n, length, chosen):
    print("\t"*length, chosen, n)
    if n < 0 or length > len(arr) or tuple(sorted(chosen)) in memo:
        print("\t"*length, "wrong")
        return 0
    elif n == 0:
        print("\t"*length, "found")
        memo.add(tuple(sorted(chosen)))
        return 1
    print("\t"*length, "going deeper")
    ans = 0
    for i in arr:
        ans += f(arr, n - i, length + 1, chosen + [i])
    return ans

def find(arr,n):
    ans = 0
    for i in arr:
        print(f"start with {i}")
        ans += f(arr, n - i, 1, [i])
    return ans

find([1, 2, 3], 7)
```
