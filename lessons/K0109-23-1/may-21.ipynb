программу, которая выводит на экран сумму цифр числа, больших 5. Если в числе нет цифр,
больших 5, требуется вывести на экран «NO». Программист написал программу
неправильно. Ниже эта программа для Вашего удобства приведена на пяти языках
программирования. 

````python
N = int(input()) #748
sum = N % 10 #8
while N > 0: 
    digit = N % 10 # 7
    if digit > 5:
        sum = digit # 7
    N = N // 10 # 7

if sum > 0:
    print(sum) 
else:
    print("NO")
````

(некоторые числа могут быть одинаковыми). Нужно написать программу, которая выводит
на экран количество делящихся нацело на 3 чисел в исходной последовательности
и максимальное делящееся нацело на 3 число. Если делящихся нацело на 3 чисел нет,
требуется на экран вывести «NO». Известно, что вводимые числа не превышают 1000.
Программист написал программу неправильно. Ниже написанная им программа для Вашего

````python
n = 4
count = 0
maximum = 999
for i in range(1, n+1):
    x = int(input()) # 3
    if x % 3 == 0:
        count += 1 # 2
        if x < maximum:
            maximum = x # 3
if count > 0:
    print(count) # 2
    print(maximum) # 3
else:
    print("NO")
````

````python
def gap(num):
    num = bin(num)[2:]
    print(num)
    
    m = 0
    c = 0
    for i in num:
        if i == '0':
            c += 1
        else:
            if c > m:
                m = c
            c = 0

    return m

    
gap(9), gap(529), gap(20), gap(15)
````
